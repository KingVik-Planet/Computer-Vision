```python
#Import cv2

import cv2
import matplotlib.pyplot as plt
import numpy as np

#Reading the Image
image = cv2.imread("images/img.png")
```

```python
# Checking if the Image is loaded successfully
if image is None:
    print("Error loading ")
else:
    plt.imshow(image)
    plt.title("Mountain Image")
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
lower_red = np.array([0, 255, 100])
upper_red = np.array([10,255,255])

mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(image, image, mask=mask)

result = cv2.cvtColor(result, cv2.COLOR_HSV2BGR)

plt.imshow(result)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
lower_red = np.array([0, 255, 100])
upper_red = np.array([0,255,2])

mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(image, image, mask=mask)

result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

plt.imshow(result)
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title ("Mountain Image")
plt.xlabel("Horizontal Grid")
plt.xlabel("Vertical Grid")
```
